```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded[next(iter(uploaded))]))
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.scatter(df['area'], df['price'])
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Relationship between Area and Price')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df.groupby('bedrooms')['price'].mean().index,
        df.groupby('bedrooms')['price'].mean().values)
plt.xlabel('Bedrooms')
plt.ylabel('Average Price')
plt.title('Average Price by Number of Bedrooms')
plt.show()
```

```python
df.groupby('prefarea')['price'].mean()
```

```python
avg_price_pref = df.groupby('prefarea')['price'].mean()

plt.figure(figsize=(6,4))
plt.bar(avg_price_pref.index.astype(str), avg_price_pref.values)
plt.xlabel('Preferred Area')
plt.ylabel('Average Price')
plt.title('Average Price by Preferred Area')
plt.show()
```
